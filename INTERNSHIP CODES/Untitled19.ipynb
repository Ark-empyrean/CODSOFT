```python
from tkinter import *
from tkinter import font, messagebox

# Function to update expression in the text entry box
def press(num):
    global expression
    expression += str(num)
    equation.set(expression)

# Function to evaluate the final expression
def equalpress():
    global expression
    try:
        total = str(eval(expression))
        equation.set(total)
        expression = ""
    except:
        equation.set("error")
        expression = ""

# Function to clear the contents of text entry box
def clear():
    global expression
    expression = ""
    equation.set("")

# Function to stop the calculator
def stop_calculator():
    messagebox.showinfo("Calculator", "Thanks for using the calculator!")
    root.destroy()

# Main code
if __name__ == "__main__":
    root = Tk()
    root.title("Calculator")

    expression = ""

    # Setting up the input field
    equation = StringVar()
    input_field = Entry(root, textvariable=equation, font=font.Font(size=20), justify='right')
    input_field.grid(columnspan=4, ipadx=8, ipady=8)

    # Button definitions here...
    # Creating buttons for the calculator
button1 = Button(root, text=' 1 ', fg='black', bg='yellow', command=lambda: press(1), height=2, width=7)
button1.grid(row=2, column=0)

button2 = Button(root, text=' 2 ', fg='black', bg='yellow', command=lambda: press(2), height=2, width=7)
button2.grid(row=2, column=1)

button3 = Button(root, text=' 3 ', fg='black', bg='yellow', command=lambda: press(3), height=2, width=7)
button3.grid(row=2, column=2)

button4 = Button(root, text=' 4 ', fg='black', bg='yellow', command=lambda: press(4), height=2, width=7)
button4.grid(row=3, column=0)

button5 = Button(root, text=' 5 ', fg='black', bg='yellow', command=lambda: press(5), height=2, width=7)
button5.grid(row=3, column=1)

button6 = Button(root, text=' 6 ', fg='black', bg='yellow', command=lambda: press(6), height=2, width=7)
button6.grid(row=3, column=2)

button7 = Button(root, text=' 7 ', fg='black', bg='yellow', command=lambda: press(7), height=2, width=7)
button7.grid(row=4, column=0)

button8 = Button(root, text=' 8 ', fg='black', bg='yellow', command=lambda: press(8), height=2, width=7)
button8.grid(row=4, column=1)

button9 = Button(root, text=' 9 ', fg='black', bg='yellow', command=lambda: press(9), height=2, width=7)
button9.grid(row=4, column=2)

button0 = Button(root, text=' 0 ', fg='black', bg='yellow', command=lambda: press(0), height=2, width=7)
button0.grid(row=5, column=1)

plus = Button(root, text=' + ', fg='black', bg='yellow', command=lambda: press("+"), height=2, width=7)
plus.grid(row=2, column=3)

minus = Button(root, text=' - ', fg='black', bg='yellow', command=lambda: press("-"), height=2, width=7)
minus.grid(row=3, column=3)

multiply = Button(root, text=' * ', fg='black', bg='yellow', command=lambda: press("*"), height=2, width=7)
multiply.grid(row=4, column=3)

divide = Button(root, text=' / ', fg='black', bg='yellow', command=lambda: press("/"), height=2, width=7)
divide.grid(row=5, column=3)

equal = Button(root, text=' = ', fg='black', bg='yellow', command=equalpress, height=2, width=7)
equal.grid(row=5, column=2)

clear = Button(root, text=' C ', fg='black', bg='yellow', command=clear, height=2, width=7)
clear.grid(row=5, column=0)

percent = Button(root, text=' % ', fg='black', bg='yellow', command=lambda: press("%"), height=2, width=7)
percent.grid(row=6, column=0)

double_zero = Button(root, text=' 00 ', fg='black', bg='yellow', command=lambda: press("00"), height=2, width=7)
double_zero.grid(row=6, column=1)

    # Adding Stop button
stop_button = Button(root, text=' Stop ', fg='red', bg='green', command=stop_calculator, height=2, width=7)
stop_button.grid(row=6, column=2)

root.geometry('350x300')
root.mainloop()
```
